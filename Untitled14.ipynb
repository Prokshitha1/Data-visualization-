```python
                                                                    # Task: Bivariate Analysis on Placement Data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Placement_Data_Full_Class.csv")

sns.set(style="whitegrid")

# ===============================
# ðŸ”· CATEGORICAL vs CATEGORICAL
# ===============================

# Crosstab Heatmap: Gender vs Status
ct = pd.crosstab(df['gender'], df['status'])
plt.figure(figsize=(6,4))
sns.heatmap(ct, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap: Gender vs Placement Status')
plt.show()

# Stacked Bar Plot: Specialisation vs Status proportions
ct2 = pd.crosstab(df['specialisation'], df['status'])
spec_status = ct2.div(ct2.sum(axis=1), axis=0)
spec_status.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Paired')
plt.title('Stacked Bar Chart: Specialisation vs Placement Status')
plt.ylabel('Proportion')
plt.show()

# ===============================
# ðŸ”¶ CATEGORICAL vs CONTINUOUS
# ===============================

# Boxplot: Degree Percentage distribution by Status
plt.figure(figsize=(8,5))
sns.boxplot(x='status', y='degree_p', data=df, palette='pastel')
plt.title('Boxplot: Degree % by Placement Status')
plt.show()

# Violin Plot: MBA Percentage distribution by Gender
plt.figure(figsize=(8,5))
sns.violinplot(x='gender', y='mba_p', data=df, palette='muted')
plt.title('Violin Plot: MBA % by Gender')
plt.show()

# Barplot: Average E-test % by Specialisation
plt.figure(figsize=(8,5))
sns.barplot(x='specialisation', y='etest_p', data=df, palette='coolwarm')
plt.title('Barplot: Avg Etest % by Specialisation')
plt.show()

# ===============================
# ðŸ”· CONTINUOUS vs CONTINUOUS
# ===============================

# Scatter Plot: Degree % vs MBA % colored by Status
plt.figure(figsize=(8,6))
sns.scatterplot(x='degree_p', y='mba_p', hue='status', data=df, palette='Set2')
plt.title('Scatter Plot: Degree % vs MBA % by Status')
plt.show()

# Jointplot: SSC % vs HSC % with regression line
sns.jointplot(x='ssc_p', y='hsc_p', data=df, kind='reg', height=7, color='purple')
plt.suptitle('Joint Plot: SSC % vs HSC % with Regression', y=1.02)
plt.show()

# Correlation Heatmap of numerical variables
plt.figure(figsize=(8,5))
numeric_cols = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Correlation Heatmap of Academic Percentages')
plt.show()
```
