```python
# Force-based Layout Visualization using NetworkX for Placement Dataset
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# ---- Step 1: Load Dataset ----
df = pd.read_csv("/content/Placement_Data_Full_Class.csv")

# ---- Step 2: Extract relationships ----
# We'll link Specialisation  Status  Gender
edges = []

for _, row in df.iterrows():
    specialization = row["specialisation"]
    status = row["status"]
    gender = row["gender"]

    # Avoid duplicates
    edges.append((specialization, status))
    edges.append((status, gender))

# Remove duplicate edges
edges = list(set(edges))

# ---- Step 3: Create a Graph ----
G = nx.Graph()
G.add_edges_from(edges)

# ---- Step 4: Compute Force-based Layout ----
pos = nx.spring_layout(G, seed=42)  # Fruchterman-Reingold algorithm

# ---- Step 5: Visualize the Graph ----
plt.figure(figsize=(8, 6))
nx.draw(
    G, pos,
    with_labels=True,
    node_color="lightgreen",
    node_size=1500,
    font_size=10,
    font_weight="bold",
    edge_color="gray"
)
plt.title("Placement Dataset: SpecialisationStatusGender Network", fontsize=14)
plt.show()
```
